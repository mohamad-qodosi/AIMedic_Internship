```python
for c in df.columns:
    if df[c].dtype == 'object':
        print('-' * 10)
        print(c)
        print(len(df[c].value_counts()))
```

```python
binary_columns = []
categorical_columns = []

for c in df.columns:
    if df[c].dtype == 'object':
        if len(df[c].value_counts()) == 2:
            binary_columns.append(c)
        if len(df[c].value_counts()) > 2:
            categorical_columns.append(c)
```

```python
print(binary_columns)
print(categorical_columns)
```

```python
df = pd.get_dummies(df, prefix=binary_columns, columns=binary_columns , drop_first=True)
df = pd.get_dummies(df, prefix=categorical_columns, columns=categorical_columns)
```

```python
df.head()
```

```python
df.columns
```

```python
y = df.price.to_numpy()
df = df.drop('price', axis=1)
X = df.to_numpy()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X2 = poly.fit_transform(X)
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression(normalize=True)
```

```python
lr_model.fit(X2, y)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
lr_model.predict(X2)
```

```python
lr_model.score(X2, y)
```

```python
h = lr_model.predict(X2)
```

```python
for i in range(X.shape[0]):
    print(h[i], y[i])
```
